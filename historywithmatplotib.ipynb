```python
import pandas as pd
%matplotlib inline
import numpy as np
from numpy.random import randn

#Create a quick dataframe
df=pd.DataFrame(randn(100,4),columns=["Mon","Tue","Wed","Thur"])
df
```

```python
#Histogram
df["Wed"].hist()
```

```python
#hISTOGRAM WITH SMALLER BINS,50,20
df["Wed"].hist(bins=20)
```

```python
# aNOTHER WAY WITHOUT GRID BARS
df["Wed"].plot(kind="hist")
```
